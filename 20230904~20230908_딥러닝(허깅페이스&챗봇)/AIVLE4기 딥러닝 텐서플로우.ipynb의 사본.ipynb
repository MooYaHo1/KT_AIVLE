# 세번째 딥러닝 - 아이리스 품종 분류

- 데이터: https://raw.githubusercontent.com/blackdew/ml-tensorflow/master/data/csv/iris.csv

```python
import tensorflow as tf
import pandas as pd

# 데이터를 준비합니다.
아이리스 = pd.read_csv("https://raw.githubusercontent.com/blackdew/ml-tensorflow/master/data/csv/iris.csv")
print(아이리스.shape)

# 원핫인코딩
아이리스 = pd.get_dummies(아이리스)
아이리스.columns

독립 = 아이리스[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
종속 = 아이리스[['품종_setosa', '품종_versicolor', '품종_virginica']]

print(독립.shape, 종속.shape)
```

```python
아이리스
```

```python
아이리스.head()
# [5.1, 3.5, 1.4, 0.2] => (4,) = [4]
# [[0, 1],
#  [1, 0]] => (2, 2) = [2, 2]
```

```python
# 모델을 준비합니다.
X = tf.keras.Input(shape=[4])
Y = tf.keras.layers.Dense(3, activation="softmax")(X)
model = tf.keras.Model(X, Y)
model.compile(loss="categorical_crossentropy", metrics="accuracy")

model.summary()
```

```python
# 데이터로 모델을 학습합니다.
model.fit(독립, 종속, epochs=500, verbose=0)
model.fit(독립, 종속, epochs=10)
```

```python
# 모델을 이용합니다.
model.predict(독립[:5])
```

```python
model.get_weights()
```

## [가정] 품종이 2 class 일때
1. 품종을 0, 1의 값으로 치환.
  - 종속변수가 1개
  - sigmoid, binary_crossentropy 사용
2. 품종을 onehot encoding => [0칼럼, 1칼럼]
  - 종속변수 2개
  - softmax, categorical_crossentropy 사용.

```python
def add(a, b):
    return a + b
# add = lambda a, b: a + b

def mul(a, b):
    return a * b
# mul = lambda a, b: a * b

print(type(add))

def func(f, a, b):
    return f(a, b)

print(func(add, 1, 2))
print(func(add, 3, 5))
print(func(mul, 3, 5))
```

```python
# 함수를 전달해서 이용하는 예시
students = [["이숙번", 80, 54], ["김건형", 90, 66], ["박종성", 85, 77]]

def get(x):
    return sum(x[1:])
# get = lambda x: sum(x[1:])

sorted(students, key=get)
sorted(students, key=lambda x: sum(x[1:]))
```

## 데코레이터를 활용한 합성함수 사용

```python
def curry(func):
    def curried(*args, **kwargs):
        if len(args) + len(kwargs) >= func.__code__.co_argcount:
            return func(*args, **kwargs)
        return (lambda *args2, **kwargs2:
            curried(*(args + args2), **dict(kwargs, **kwargs2)))
    return curried

@curry
def foo(a, b, c):
    return a - b - c
```

```python
print(foo(1, 2, 3))
print(foo(1)(2))
print(foo(1)(2)(3))
```

# 네번째 딥러닝 - 진짜 딥뉴럴넷

```python
import tensorflow as tf
import pandas as pd

# 데이터를 준비합니다.
보스턴 = pd.read_csv("https://raw.githubusercontent.com/blackdew/ml-tensorflow/master/data/csv/boston.csv")
보스턴.columns

독립 = 보스턴[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'b', 'lstat']]
종속 = 보스턴[['medv']]

print(독립.shape, 종속.shape)
```

```python
# 모델을 준비합니다.
X = tf.keras.Input(shape=[13])
H = tf.keras.layers.Dense(128, activation='swish')(X)
Y = tf.keras.layers.Dense(1)(H)
model = tf.keras.Model(X, Y)
model.compile(loss='mse')

model.summary()
```

```python
model.fit(독립, 종속, epochs=1000)
```
