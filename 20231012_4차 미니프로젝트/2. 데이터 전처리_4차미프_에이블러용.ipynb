# 1. 데이터 전처리

#### [실습문제1] 열 추가
* 기존 '기준일ID'열에서 날짜 데이터를 추출하여 새로운 열로 추가 합니다.
    - pd.to_datetime() 함수를 사용해 '기준일ID'열을 datetime 형식으로 변경하겠습니다.
    - year, month, day 열로 기존 데이터 프레임에 연월일 정보를 저장합니다.
    - 데이터 프레임 열에 대해 날짜 정보를 얻을 때에는 **dt.year**, **dt.month**, **dt.day** 속성을 사용합니다.

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
train_df['기준일ID'] = pd.to_datetime(train_df['기준일ID'])
train_df['year'] = train_df['기준일ID'].dt.year
train_df['month'] = train_df['기준일ID'].dt.month
train_df['day'] = train_df['기준일ID'].dt.day
train_df.head()
```

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.

test_df['기준일ID'] = pd.to_datetime(test_df['기준일ID'])
test_df['year'] = test_df['기준일ID'].dt.year
test_df['month'] = test_df['기준일ID'].dt.month
test_df['day'] = test_df['기준일ID'].dt.day
display(test_df.head())
display(test_df.tail())
```

#### [실습문제3] 시프트 & 롤링
* shift()를 활용하여 이동할 행 수를 지정합니다.
* rolling()을 활용하여 추가할 이동평균 수를 지정합니다.

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
train_df['총생활인구수_24h'] = train_df['총생활인구수'].shift(-24)
train_df.head(25)
```

```python
train_df['weekday'] = train_df['기준일ID'].dt.weekday
train_df['총생활인구수_3h'] = train_df['총생활인구수'].rolling(3, min_periods = 1).mean()
```

```python
train_df['총생활인구수_6h']=train_df['총생활인구수'].rolling(6, min_periods = 1).mean()
```

```python
train_df['총생활인구수_y24h'] = train_df['총생활인구수'].shift(24)
```
