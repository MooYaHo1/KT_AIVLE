* 모델링 라이브러리 불러오기

```python
from ultralytics import YOLO
```

### 1) 데이터셋에서 제공하는 모델 사용하기

* 데이터셋 제공 모델 불러오기

### 2) 모델 구조만 빌려와서 사용해보기

* 모델 구조 선택하기

```python
model_scratch = YOLO(model='yolov8n.yaml', task='detect')
```

* 모델 학습하기

```python
model_scratch.train(data='/content/test_project-2/data.yaml',
                    epochs=10,
                    patience=5,
                    pretrained=False,
                    seed=2023,
                    )
```

* 예측해보기

```python
model_scratch.predict(source='https://www.cj-ilbo.com/news/photo/202206/1082109_996141_4325.jpg',
                      conf=0.7,
                      iou=0.1,
                      save=True,
                      line_width=2
                      )
```

* 모델 구조 및 사전 학습 가중치 선택하기
