```python
#read_csv 재실행. 단, header값을 적절히 지정
data = pd.read_csv('3. seoul_people.csv', sep='\t', encoding='cp949', header = 2)
data.head()
```

```python
data1.drop(0, axis=0, inplace=True)
data1
```
