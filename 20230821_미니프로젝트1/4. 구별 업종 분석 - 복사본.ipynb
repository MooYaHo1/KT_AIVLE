#### [4.2.2] 원하는 데이터 형태로 저장하기

* head, tail 등을 살펴본 결과를 토대로 header 값을 적절히 지정하여 데이터를 다시 불러오고 필요한 컬럼만 저장하세요

```python
#read_csv 재실행. 단, header값을 적절히 지정
seoul_business = pd.read_csv('4. business_type.csv', encoding='UTF8', header=1)
seoul_business.head()
```

```python
seoul_business.columns
```

```python
# 다시 불러온 데이터프레임에서 여러분 각자가 사용하고자 하는 컬럼만 가져오세요.
seoul_business = seoul_business[['자치구', '동',  '한식 일반 음식점업.1', '부동산 중개 및 대리업.1', '두발미용업.1', '커피전문점.1', 
       '기타주점업.1', '일반 교과 학원.1', '한식 육류요리 전문점.1']]
seoul_business.head()
```

```python
# 여러분 각자가 사용하고자 하는 형태로 데이터를 정리합니다. 
# 예를 들어, 동 컬럼의 값이 소계인 항목들만 골라내 보겠습니다. 
# 위와 같이 동 컬럼의 소계를 구하면 자치구 별 값이 되므로 이후에 동 컬럼은 삭제하겠습니다. 결과적으로 자치구 별 값을 얻게 됩니다.
data1 = seoul_business.loc[seoul_business['동'] == '소계']
data1
```

```python
data1.drop('동', axis=1, inplace=True)
```

```python
data1.reset_index(drop=True)
```

```python
# 위에서 여러분 각자가 사용하고자 하는 형태로 정리한 해당 데이터프레임을 csv 파일('df_seoul_business.csv')로 저장하세요.
data1.to_csv('df_seoul_business.csv', index=False, encoding='euc-kr')
```

# 4.3.데이터 분석하기
+ 데이터의 형태를 살펴보고 다양한 분석기법을 통해 모델링에 적합하도록 정제요소를 선별하세요.
  * 데이터들의 패턴 탐색
  * 변수들간의 관계 파악

```python
# 시각화 한글폰트 설정
import seaborn as sns
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')
sns.set(font="Malgun Gothic",#"NanumGothicCoding", 
        rc={"axes.unicode_minus":False}, # 마이너스 부호 깨짐 현상 해결
        style='darkgrid')        
```

#### [4.3.1] 데이터 분포 알아보기_1
* '자치구' 등 여러 변수를 기준으로 시각화 해보고 인사이트를 도출해보세요.

```python
# 아래에 코드를 작성하고 결과를 확인합니다.
data1.info()
```

```python
data2 = data1.copy()
```

```python
data2.info()
```

```python
data2['한식 일반 음식점업.1'] = data2['한식 일반 음식점업.1'].str.replace(',','').astype('int64')
```

```python
data2['한식 일반 음식점업.1']
```

```python
columns=['부동산 중개 및 대리업.1','두발미용업.1','커피전문점.1','기타주점업.1','일반 교과 학원.1','한식 육류요리 전문점.1']
for column in columns:
    data2[column] = data2[column].str.replace(',','').astype('int64')
print(data2.info())
print(data2.head())
```

```python
data2.head()
```

```python
plt.figure(figsize=(16,10))

plt.subplot(3,3,1)
sns.barplot(x='자치구', y='한식 일반 음식점업.1', data=data2)
plt.xticks(rotation=90)

plt.subplot(3,3,2)
sns.barplot(x='자치구', y='부동산 중개 및 대리업.1', data=data2)
plt.xticks(rotation=90)

plt.subplot(3,3,3)
sns.barplot(x='자치구', y='두발미용업.1', data=data2)
plt.xticks(rotation=90)

plt.subplot(3,3,4)
sns.barplot(x='자치구', y='커피전문점.1', data=data2)
plt.xticks(rotation=90)

plt.subplot(3,3,5)
sns.barplot(x='자치구', y='기타주점업.1', data=data2)
plt.xticks(rotation=90)

plt.subplot(3,3,6)
sns.barplot(x='자치구', y='일반 교과 학원.1', data=data2)
plt.xticks(rotation=90)

plt.subplot(3,3,7)
sns.barplot(x='자치구', y='한식 육류요리 전문점.1', data=data2)
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

```python
# 위 차트를 통해 알게된 사실을 정리해봅시다.
# 1. 모든 업종이 강남구에 월등히 많다
# 2. 주점이 많은 구인 마포구와 강남구에는 대중교통 이용자가 많을 것 같다.
# 3. 학원이 많은 구인 양천구와 강남구는 학생들이 많이 대중교통을 이용할 것 같다.
```
