```python
from sklearn.metrics import *
# 각 모델을 선언하고 학습
models = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "SVM": SVC(),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Naive Bayes": GaussianNB(),
    "SGD Classifier": SGDClassifier(),
    "AdaBoost": AdaBoostClassifier(),
    "Gradient Boosting": GradientBoostingClassifier()
}

# 모델 학습 및 예측
for model_name, model in models.items():
    model.fit(x_tr_tfidfv, y_tr)
    y_pred = model.predict(x_val_tfidfv)

    # 결과 출력
    accuracy = accuracy_score(y_val, y_pred)
    print(f"Model: {model_name}")
    print(f"Accuracy: {accuracy:.4f}")
    print()
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
model = SGDClassifier()

# 하이퍼파라미터 그리드 정의
param_grid = {
    'loss': ['log_loss', 'modified_huber'],
    'penalty': ['l2', 'l1'],
    'alpha': [0.0001, 0.001, 0.01],
    'max_iter': [100, 500, 1000],
}

# 그리드 탐색 수행
grid_search = GridSearchCV(model, param_grid, cv=5,verbose=2)
grid_search.fit(x_tr_tfidfv, y_tr)

# 최적 하이퍼파라미터 출력
print("Best Hyperparameters: ", grid_search.best_params_)

# 최적 모델로 예측
best_model = grid_search.best_estimator_
y_pred = best_model.predict(x_val_tfidfv)

# 정확도 및 분류 보고서 출력
accuracy = accuracy_score(y_val, y_pred)
report = classification_report(y_val, y_pred)

print(f'Accuracy: {accuracy}')
print('Classification Report:\n', report)
```

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.linear_model import SGDClassifier

# 모델 초기화
model = SGDClassifier()

# K-Fold Cross-Validation 객체 생성
kfold = KFold(n_splits=5)  # 5개의 폴드로 나눔 (K=5)

# 교차 검증 수행
scores = cross_val_score(model, x_tr_tfidfv, y_tr, cv=kfold)

# 교차 검증 결과 출력 (각 폴드의 정확도)
print("Cross-Validation Scores:", scores)

# 평균 정확도 출력
print("Mean Accuracy:", scores.mean())
```
