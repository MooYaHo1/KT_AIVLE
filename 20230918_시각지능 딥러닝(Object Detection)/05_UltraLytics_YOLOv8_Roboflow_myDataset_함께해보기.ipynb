* 모델링 라이브러리 불러오기

### 1) 모델 구조만 빌려와서 사용해보기

* 모델 구조 선택하기

* 모델 학습하기

### 2) 모델 구조에 사전 학습 가중치도 가져와서 사용해보기

* 모델 구조 및 사전 학습 가중치 선택하기

```python
model = YOLO(model='yolov8n.pt', task='detect')
```

* 모델 학습하기

```python
model.train(data='/content/face_classification-1/data.yaml',
            epochs=100,
            patience=20,
            pretrained=True,
            )
```

* 예측해보기

```python
path='https://cdn.iworldtoday.com/news/photo/201509/175136_48996_4817.jpg'
```

```python
results = model.predict(source=path,
                         save=True,
                         line_width=2
                         )
```
